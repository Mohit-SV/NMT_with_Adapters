```python
import pandas as pd
```

```python
# FOR BRAZILLIAN PORTUGUESE DATASET
```

```python
# merge all data
df_train = pd.read_csv('brazil_train.csv')
df_test = pd.read_csv('brazil_test.csv')
df_val = pd.read_csv('brazil_validation.csv')

df = pd.concat([df_train, df_test, df_val], ignore_index=True, axis=0)

df_en = pd.concat( [df.iloc[:, 2], df.iloc[:, 4], df.iloc[:, 6], df.iloc[:, 8] ], ignore_index=True, axis=0)
df_pt = pd.concat( [df.iloc[:, 3], df.iloc[:, 5], df.iloc[:, 7], df.iloc[:, 9] ], ignore_index=True, axis=0)

```

```python
df_en = pd.concat( [df.iloc[:, 2], df.iloc[:, 4]], ignore_index=True, axis=0)
df_pt = pd.concat( [df.iloc[:, 3], df.iloc[:, 5]], ignore_index=True, axis=0)

df_merged = pd.concat([df_en, df_pt], ignore_index=True, axis=1)
df_merged.columns = ['en', 'pt']


df_merged = df_merged[df_merged['en'].notna()]
df_merged
```

```python
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt


nr_sentence = len(df_merged)

sentence_lengths_en = df_merged['en'].apply(lambda n: len(n.split()))

average_length_en = (sentence_lengths_en.sum() / (len(sentence_lengths_en)))

nr_words_en = sentence_lengths_en.sum()

unique_words_en = Counter()
df_merged['en'].str.lower().str.split().apply(unique_words_en.update)


print('****ENGLISH****')
print(f'Nr of sentences English: {nr_sentence}')
print(f'Nr of words: {nr_words_en}')
# print(f'Unique words: {unique_words_en}')
print(f'Average sentence length: {average_length_en}')

length_list = []

for sentence in df_merged['en']:
    length_list.append(len(sentence.split()))
sns.histplot(length_list, bins = 50)
plt.title('English sentence length distribution')
plt.xlabel('Sentence length')
plt.show()


#####################################################Brazilian Portuguese
sentence_lengths_af = df_merged['pt'].apply(lambda n: len(n.split()))

average_length_af = (sentence_lengths_af.sum() / (len(sentence_lengths_af)))

nr_words_af = sentence_lengths_af.sum()

unique_words_af = Counter()
df_merged['pt'].str.lower().str.split().apply(unique_words_af.update)

print('****BRAZILIAN PORTUGUESE****')
print(f'Nr of sentences Brazilian Portuguese: {nr_sentence}')
print(f'Nr of words: {nr_words_af}')
# print(f'Unique words: {unique_words_af}')
print(f'Average sentence length: {average_length_af}')

for sentence in df_merged['pt']:
    length_list.append(len(sentence.split()))
sns.histplot(length_list, bins = 50)
plt.title('Brazilian Portuguese sentence length distribution')
plt.xlabel('Sentence length')
plt.show()
```

```python

# Simple Way to Read TSV Files in Python using split
ans = []
 
# open .tsv file
with open("port_en.tsv") as f:
   
  # Read data line by line
    for line in f:
     
        # split data by tab
        # store it in list
        l=line.split('\t')

        # append list to ans
        ans.append(l)
 
        
df2 = pd.DataFrame(ans)
df2
```
